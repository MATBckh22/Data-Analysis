```python
data.drop(['Unnamed: 0'], axis=1, inplace=True)
data.head()
```

### 2.分析price的分布，绘制直方图

```python
price_columns_new= [col for col in data.columns if 'Price' in col]

import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")

n = len(price_columns_new)
ncols = 3
nrows = n // ncols + (n % ncols > 0)


fig, axes = plt.subplots(nrows, ncols, figsize=(15, nrows*4))
fig.tight_layout(pad=5.0)


for i, col in enumerate(price_columns_new):
    ax = axes.flat[i]
    sns.histplot(data[col], kde=True, ax=ax)
    ax.set_title(col.replace('_', ' ').title(), fontsize=14)
    ax.set_xlabel('Price', fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)


for j in range(i+1, nrows*ncols):
    axes.flat[j].set_visible(False)

plt.show()
```

### 3.时序图

```python
#画出时序图
import matplotlib.pyplot as plt
import seaborn as sns


if 'Date' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Sorting data by date to ensure correct trend visualization
new_data_sorted = data.sort_values('Date')

# Plotting price trends over time
plt.figure(figsize=(15, 20))

for i, col in enumerate(price_columns_new):
    plt.subplot(len(price_columns_new), 1, i+1)
    plt.plot(new_data_sorted['Date'], new_data_sorted[col], label=col)
    plt.title(f'{col.replace("_", " ")} Over Time')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()

plt.tight_layout()
plt.show()
```

### 4.分析各个price之间的关系，画出相关系数矩阵

```python
# Calculate the correlation matrix for price variables
correlation_matrix = data[price_columns_new].corr()

# Plotting the heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Price Variables')
plt.show()
```

### 5.选取相关性比较大的数据进行格兰杰因果检验

```python
#这里以Microsoft和Apple为例
from statsmodels.tsa.stattools import grangercausalitytests


microsoft_prices = new_data_sorted['Microsoft_Price']
apple_prices = new_data_sorted['Apple_Price']


time_series_data = pd.concat([microsoft_prices, apple_prices], axis=1)#在这里可以换成别的变量
time_series_data.columns = ['Microsoft_Price', 'Apple_Price']


max_lags = 5
granger_test_results = grangercausalitytests(time_series_data, maxlag=max_lags, verbose=True)
```

在滞后期1时，p值接近0.05，表明苹果股价对微软股价没有很好的预测能力。
在滞后期2至5时，p值显著小于0.05，这表明在这些滞后期内，苹果股价对微软股价具有显著的预测能力。
同理也可以根据其他数据进行因果分析


### 6.使用pca进行分析

```python

from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import StandardScaler

fa_data = data[price_columns_new]

#标准化
scaler = StandardScaler()
fa_data_scaled = scaler.fit_transform(fa_data)

# PCA
pca = PCA(n_components=len(price_columns_new))
pca.fit(fa_data_scaled)

#方差解释率
explained_variance_ratio = pca.explained_variance_ratio_

#累计方差解释率
cumulative_explained_variance = explained_variance_ratio.cumsum()

explained_variance_ratio, cumulative_explained_variance
```

第一个主成分解释了最大比例的方差（约42.53%），表明第一个主成分对price的影响最大

### 7.因子分析

```python
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity

#选择price列
fa_data = data[price_columns_new]

#进行因子分析
kmo_all, kmo_model = calculate_kmo(fa_data)
bartlett_sphericity_test_stat, p_value = calculate_bartlett_sphericity(fa_data)

#输出结果
kmo_model, bartlett_sphericity_test_stat, p_value
```

Kaiser-Meyer-Olkin (KMO) 测度：0.7949001622166385<br>

KMO 测度用于评估变量之间的偏相关性小于它们的整体相关性的程度，作为因子分析适用性的一个指标。KMO值范围从0到1，较高的值（通常大于0.6）表明数据集适合进行因子分析。在这个例子中，KMO值为0.79，表明变量间的共通性高，因子分析是合适的。<br>

Bartlett的球形度检验统计量：32207.785042481435<br>

Bartlett的球形度检验用于检验数据是否在多变量正态分布下具有等方差性，即所有变量的相关系数矩阵是单位矩阵。这个检验的统计量很高，表明变量之间有足够的关联，适合进行因子分析。<br>

Bartlett检验的p值：0.0<br>

这个极低的p值（实际上可能是接近于0但因为数值精度显示为0）表明我们可以拒绝原假设（即相关矩阵是单位矩阵，变量间无关联）。因此，数据中的变量间存在相关性，支持进行因子分析。
