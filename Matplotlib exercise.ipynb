```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,50)
y = [value * 3 for value in x]
plt.plot(x, y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("a line")
plt.show()
```
